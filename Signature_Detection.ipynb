```python
from google.colab import files
import os
import zipfile

# Upload the zip file
uploaded = files.upload()

# Get the uploaded file name
dataset_path = next(iter(uploaded))  # Automatically get the uploaded file's name
extract_dir = "/content/dataset"

# Create directories for extraction
os.makedirs(extract_dir, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(dataset_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Dataset extracted to {extract_dir}")

# Define paths for train and validation images/labels
train_images_path = os.path.join(extract_dir, "train", "images")
train_labels_path = os.path.join(extract_dir, "train", "labels")
val_images_path = os.path.join(extract_dir, "valid", "images")
val_labels_path = os.path.join(extract_dir, "valid", "labels")

# Define the classes (add more classes if needed)
classes = ["signature"]

# Create the data.yaml content
data_yaml_content = f"""
train: {train_images_path}
val: {val_images_path}
nc: {len(classes)}
names: {classes}
"""

# Save data.yaml
data_yaml_path = os.path.join(extract_dir, "data.yaml")
with open(data_yaml_path, "w") as yaml_file:
    yaml_file.write(data_yaml_content)

print(f"data.yaml created with content:\n{data_yaml_content}")

# Now you can start training YOLO
!yolo task=detect mode=train data={data_yaml_path} model=yolov8n.pt epochs=50 imgsz=640

```
